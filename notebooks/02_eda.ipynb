```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/sentiment_project/processed_reviews_50k.csv')
```

```python
df['rating'].value_counts(normalize=True).plot(kind='bar', title='Rating Distribution')
```

```python
df['sentiment'].value_counts(normalize=True).plot(kind='bar', title='Sentiment Distribution')
```

```python
class_counts = df['label'].value_counts()

class_percentages = (class_counts / len(df)) * 100

print(class_percentages)
```

```python
sns.countplot(x='label', data=df)
plt.title("Distribution of Sentiment Classes")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
df['text_length'] = df['full_text'].apply(lambda x: len(str(x).split()))

# Average length per sentiment
df.groupby('label')['text_length'].mean()

# Histogram of text lengths
sns.histplot(data=df, x='text_length', bins=50, hue='label', element="step")
plt.title("Text Length Distribution by Sentiment")
plt.show()
```
